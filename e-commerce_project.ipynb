```python
# Пометка для себя: 
# Перед проведением когортного анализа определяем четыре параметра:

# Признак формирования когорты — действие, которое объединяет людей в группу: повторная покупка и цена покупки
# Размер когорты — временной интервал: месяц
# Отчетный период — время исследования поведения групп: год
# Анализируемый ключевой показатель: Retention Rate = (количествo клиентов на конец периода - количество новых клиентов
# за период)/количество клиентов на начало периода
```
